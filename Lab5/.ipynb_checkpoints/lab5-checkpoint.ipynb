```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt

df = pd.read_excel('DataSet//kr1.xlsx')

data_row = df.iloc[1]  

image_filename = data_row['file name']

color_dict = {
    "red": (0, 0, 255),
    "blue": (255, 0, 0),
    "yellow": (0, 255, 255),
    "cian": (255, 255, 0),
    "magenta": (255, 0, 255),
}

glasses_color_str = data_row['glasses color'] 

glasses_color = color_dict.get(glasses_color_str.lower(), (0, 0, 0))
line_width = data_row['line width']
image_size = tuple(map(int, data_row['image size'].split('x'))) 


def add_glasses_to_face_and_resize(image_path, output_path, image_size):

    image = cv2.imread(image_path)
    
    image = cv2.resize(image, image_size)
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
    
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = image[y:y+h, x:x+w]
        
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        if len(eyes) == 2:
            eyes = sorted(eyes, key=lambda ex: ex[0])
            (ex1, ey1, ew1, eh1), (ex2, ey2, ew2, eh2) = eyes

            eye1_center = (x + ex1 + ew1 // 2, y + ey1 + eh1 // 2)
            eye2_center = (x + ex2 + ew2 // 2, y + ey2 + eh2 // 2)
            radius = max(ew1, ew2) // 2
            
            
            cv2.circle(image, eye1_center, radius, glasses_color, line_width) 
            cv2.circle(image, eye2_center, radius, glasses_color, line_width)
            
            
            bridge_y = (eye1_center[1] + eye2_center[1]) // 2
            cv2.line(image, (eye1_center[0] + radius, bridge_y), (eye2_center[0] - radius, bridge_y), glasses_color, line_width)
            
           
            temple_length = 60  
            cv2.line(image, (eye1_center[0] - radius, eye1_center[1]),
                     (eye1_center[0] - radius - temple_length, eye1_center[1]), glasses_color, line_width)
            cv2.line(image, (eye2_center[0] + radius, eye2_center[1]),
                     (eye2_center[0] + radius + temple_length, eye2_center[1]), glasses_color, line_width)

    cv2.imwrite(output_path, image)

input_image_path = f'DataSet//Images//{image_filename}'
output_image_path = 'output_image_with_glasses_and_resized.jpg'
add_glasses_to_face_and_resize(input_image_path, output_image_path, image_size)

input_image = cv2.imread(input_image_path)
output_image = cv2.imread(output_image_path)

input_image_rgb = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
output_image_rgb = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image_rgb)
plt.title("Вхідне зображення")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(output_image_rgb)
plt.title("Вихідне зображення з окулярами")
plt.axis("off")

plt.show()
```
