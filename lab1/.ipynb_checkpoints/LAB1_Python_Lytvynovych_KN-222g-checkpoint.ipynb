## Варіант 1. Дослідження втрат літаків
1. Створіть функцію, яка приймає назву файлу та назву стовпця як параметр і повертає масив numpy зі значеннями стовпця.
2. Використовуючи функцію з попереднього завдання, створіть масив numpy зі значень стовпця «літак».
3. Знайдіть добові втрати літаків.
4. Знайдіть 3 найбільші добові втрати літаків.
5. Визначте, скільки літаків було збито влітку 2022 року.
6. Знайдіть середнє значення втрат авіації за останні 300 днів війни.
7. Створіть графік втрат літаків за перший рік війни (починаючи з першої дати в наборі даних)

```python
import csv 
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

і Function to load dates from the specified column in a file and return them as a numpy array.
def get_all_dates(file_name, dates_title):
    total_dates = []
    with open(file_name) as file:
        csvFile = csv.reader(file)
        header_row = next(csvFile)  
        date_col_index = header_row.index(dates_title) 

        for row in csvFile:
            row_date = row[date_col_index] 
            total_dates.append(row_date) 

    dates = np.array(total_dates)  
    return dates

# 5. Function to calculate total aircraft losses between two given dates
def find_loses(start, end):
    start_date = datetime.strptime(start, '%Y-%m-%d') 
    end_date = datetime.strptime(end, '%Y-%m-%d') 
    total_aircraft = []
    with open('russia_losses_equipment.csv') as file:
        csvFile = csv.reader(file)
        header_row = next(csvFile)
        date_col_index = header_row.index('date')  
        aircraft_col_index = header_row.index('aircraft') 
        for row in csvFile:
            row_date = datetime.strptime(row[date_col_index], '%Y-%m-%d')  
            if (start_date <= row_date <= end_date): 
                total_aircraft.append(int(row[aircraft_col_index]))  
        total_loses = np.diff(total_aircraft) 
        result = abs(np.sum(total_loses))  
        return result

# 1 Function to load a column from the CSV file as a numpy array
def load_column_as_numpy(filename, column_name):
        with open(filename, newline='', encoding='utf-8') as csvfile:
            reader = csv.reader(csvfile)
            header = next(reader)  
            column_index = header.index(column_name)  
            data = [row[column_index] for row in reader if row[column_index]] 
            return np.array(data, dtype=float)  
        
# 2. Main function to analyze aircraft losses
def analyze_plane_losses(filename, column_name):

    cumulative_losses = load_column_as_numpy(filename, column_name) 

    # 3. Calculate daily losses
    daily_losses = np.abs(np.diff(cumulative_losses))
    print("Daily aircraft losses:")
    print(daily_losses)

    # 4. Find the top 3 highest losses
    top_3 = np.sort(daily_losses)[-3:]
    print("\nTop 3 daily aircraft losses:")
    print(top_3)

    # 5. Calculate summer losses 
    summer_loses = find_loses('2022-06-01','2022-08-31')
    print("Aircraft losses during summer 2022:")
    print(summer_loses)

    # 6. Calculate average losses over the last 300 days
    avg_last_300_days = np.mean(daily_losses[-300:])
    print("\nAverage daily aircraft losses over the last 300 days:")
    print(avg_last_300_days)

    # 7. Plot the losses over time
    dates = np.sort(get_all_dates(filename, 'date')[1:])
    plt.figure(figsize=(16, 9), dpi=200)
    plt.xticks(np.arange(0, len(dates[:365]), 30), labels=dates[np.arange(0, len(dates[:365]), 30)], rotation=35) 
    plt.plot(dates[:365], daily_losses[:365], linestyle='--', label='Daily aircraft losses')  
    plt.title('Daily aircraft losses during the first year of the war') 
    plt.xlabel('Dates') 
    plt.ylabel('Number of aircraft')  
    plt.legend()
    plt.grid(True, which='both', color='black', linestyle='-', linewidth=0.3) 
    
    plt.savefig('plane_losses.png')
    plt.show()

    print("\nPlot saved as 'plane_losses.png'")


analyze_plane_losses('russia_losses_equipment.csv', 'aircraft')
```
